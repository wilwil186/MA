# ¿Cuándo usar regresión logśitica?

## Ventajas 

- Fácil de implementar
- Coeficientes interpretables
- Inferencia de la importancia de cada caracterśtica.
- Clasificación en porcentajes. 
- Excelentes resusltados con datasets linealmente separables. 

## Desventajas

- Asume linelidad entre las variables dependientes. 
- Overfitting sobre datasets de alta dimensionalidad. 
- Le afecta la multicolinalielidad de variables. 
- Mejores resultados con datasets grandes. 

## ¿Cuándo usuarla?
- Sencillo y rápido 
- Probabilidades de ocurrencia de un evento. 
- Dataset linealmente separable. 
- Datasets grandes. 
- Datasets balanceados.

# Carga y preprocesamiento de datos multiclase


```python
# import packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('/content/drive/MyDrive/Notas/platzi/Ruta  Machine Learning con Python/Regresión logística con Python Y sckit-learn/Dry_Bean.csv')
```

```python
df
```

# Modelo multiclase 

```python
def logistic_model(C_,solver_ , multiclass_):
    logistic_regression_model = LogisticRegression(random_state=42, solver= solver_, multi_class= multiclass_, n_jobs=-1, C=C_ )
    return logistic_regression_model
```

```python
model = logistic_model(1,'saga','multinomial')
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(confusion_matrix(y_test,predictions))
print(accuracy_score(y_test,predictions))
```
